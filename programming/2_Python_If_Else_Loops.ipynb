## If-then-else


### If

You can control the flow of your programs with a flow control statement. One of the most common and simplest ways to do so is by using an if statement. 

The general syntax of an if statement in Python is:

```
if condition:
    <do_this_if_condition_true>
```

The first part of the if statement is called a condition. It starts with the keyword `if` and ends with a colon `:`. 
The condition of an if statement always has to evaluate to a boolean value. Depending on the outcome the if statement decides whether the indented code block, also called the clause, will be executed or not. If the condition evaluates to `True` the indented block will run, if it evaluates to `False` the code will be skipped. 

> **Note:** Do you notice how the if statement, after the condition, ends in a colon `:`. This is the way that Python declares the start of an indentation (empty spaces) block.


`````python
# Run the cell and see if the output is what you would expect.
# Can you change the value of x so that the condition evaluates to False?
x = 7

if x > 5:
    print(x)
`````

`````python
# It also works with strings! Change the content of the variable school to somthing else and see what happens.
school = "neuefische"

if school == "neuefische":
    print("Hooray!")
`````

### Else

Every `if` clause can be followed by an (optional) `else` clause. As you can see in the following example the `else` clause does not have a condition. That's because it will only execute when the `if` clause evaluates to `False` and therefore it doesn't need an own condition. 

```
if condition:
    <do_this_if_condition_true>
else:
    <do_this_if_condition_false>
````


`````python
# Feel free to change the number of x to see what happens.
x = 7 

if x > 10: 
    print(x)
else:
    print("The number is to small.")
`````

### Elif

Using if and else is great but it's still very limited: Only one of the clauses will execute. 

You will encounter the situation where you want one of *many* conditions to be executed. Therefore you can use the `elif` statement. `elif` (a combination of the words else and if) always follows an `if` or another `elif` statement. It works like the `if` statement and checks a specific condition. You can add as many `elif`s as you want to your if-then statement.

### User input

We can combine our if-then statements with asking a user for input. It's pretty straightforward to ask for user input in Python. We can use the built-in function `input()` for that. In the parentheses of the function we can write a message or request as a string (remember the quotes?). Since everything we read from the user is automatically stored as a string we have to manually change the typ of the input to an integer in order to compare it to other numbers. As you can see we can do this using `int()`. 

`````python
# Asking the user for a number 
user_input = int(input("Please enter a number: "))

if user_input > 0:
    print("Your number is bigger than 0.")
elif user_input < 0:
    print("Your number is smaller than 0.")
else:
    print("Your number is 0.")
`````

## while loops
First we start off with a *while* loop. It works as follows:

    while (<condition>):
        do-something
        
So, as long as the <condition> is true, we do-something.
Let's illustrate the functionality of while loops with an example:
We write a program that sequentially prints out all the numbers from 1 to 10.

`````python
number = 1
while(number <= 10):
    print(number)
    number = number + 1
`````

## for loops

for-loops are an other type of loops. It works very similar, but goes through elements of a list:

    for x in list:
        do_something
    
We read it like: "do-somethings for every item x in the list".
So here is an example:


`````python
days = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']

for x in days:
    print( "It's", x)
`````

For numbers it's even easier as we don't have to write down all numbers from 1 to 10, but we can use the `range()`function:


`````python
for x in range(1,10):
    print(x)
`````

>**Note:** The last number of the range (10) **is not** included!

You see, with this for-loop we can do very similar tasks as before with the while-loop. Also here we can add numbers from 1 to 10:

`````python
total_sum = 0
for number in range(1,11):
    total_sum = total_sum + number
print(total_sum)
`````

We can even loop through two lists and create all possible combinations of the two of them. This is called **nested loops**.

`````python
days = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
times = ['7', '15', '20']

for day in days:
    for time in times:
        print("It's", day, ", ", time, "o'clock.")
`````

## more control in loops

We can even have more control over our loops. 
For example if we want to skip certain elements in a list, we can use `continue`. For example this kind of clock we built above doesn't need to tell us the time on saturday. To implement this, we can check wether day is saturday, and if it is, we can continue our loop without printing anything.

`````python
days = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
times = ['7', '15', '20']

for day in days:
    if day == 'saturday':
        continue
    for time in times:
        print("It's", day, ", ", time, "o'clock.")
`````

Another, more aggressive method to control the loops is `break`. If this condition is met, the loop is stoped entirely.
If our alarm should not work after saturday, then we could use `break`. You will see the difference, as our alarm is not working any further after saturday. On sunday we can now sleep as long as we want too!

`````python
days = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
times = ['7', '15', '20']

for day in days:
    if day == 'saturday':
        break
    for time in times:
        print("It's", day, ", ", time, "o'clock.")
`````
