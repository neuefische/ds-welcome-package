# Introduction to Google Colab

**Welcome to your first colab notebook and welcome to programming!**

Google Colab (Colaboratory) is a data analysis tool that allows you to write and run code, as well as write simple text (like this) to accompany your code with an explanation. And it all happens in a document within Google Drive. We use it here because it has the great advantage that no configuration is required beforehand.
 Therefore, we can get started right away. 
The first thing to know about colab is that any cell block [of code] can be executed by pressing Shift+Enter. Alternatively you can press the "play" button left to the cell. 


Colab-first-code.png

Try to execute the following code block:

```
# Your first line of code in Colab
print('This is code being run.')
```

As you can see, running the code above printed text to the screen. You will learn later that "print" is an example of a *function*. Functions are blocks of reusable code that execute a particular task. 

## Cell Types

As already mentioned in Colab we can write code as well as text.

You might have started to notice that the code blocks have little notes that say **[ ]** before you run them, and then are filled with a number after you run them. This is important, as it tells you what order cell blocks were run. (Which can sometimes affect how a program runs.)

You may also notice that other cell blocks, such as this one, do not have the **[ ]** label as with the code blocks. This is because this cell block is formatted as **Markdown** rather then code. The details of Markdown are not important here but just know you can use markdown cell blocks to display text. They are really useful for embedding notes and explanations in your Jupyter Notebook. You can add a new code block or text block by pressing "+ Code" or "+ Text" on the menu at the top or directly below your cell of interest in the document.
Try it out!

## Python Comments

Notice the green text after the **#** in the code cell?  
Anything that follows a **#** in Python is a comment and is not actually executed.  
This is a useful feature to annotate your code for yourself and others so that your code is easy to read.

## Autocomplete and getting Help


A great feature of Colab is the autocomplete function. When you start typing code, you can see a subset of the available commands. You can use "tap" or "enter" to auto-complete the first command, otherwise you can use the arrow keys to scroll through all the commands and press "enter" to complete the command.





Colab-autocomplete.png


## Call Docstrings

Even better, you can even see how a method works by invoking its **docstring**!  
You can do this by writing **?** after the method and executing the cell.




Colab-help.png

Try it out!

```
print?
```

Alternatively, **within the parentheses of a method** the docstring will appear automatically.

Try it out!

## Stop Execution

Long running python processes can be interrupted. Run the following cell and select **Runtime -> Interrupt execution** (hotkey: *Cmd/Ctrl-M I*) to stop execution.

```
import time
print("Sleeping")
time.sleep(30) # sleep for a while; interrupt me!
print("Done Sleeping")
```
